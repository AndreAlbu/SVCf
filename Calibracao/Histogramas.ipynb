```python
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread("1.png")
azul, verde, vermelho = cv.split(img)

#plt.imshow(img)
fig = plt.figure(figsize = (20, 8))

ax1 = fig.add_subplot(131)
ax1.hist(azul.ravel(), 256, [3, 256])
plt.title("Histograma cor azul")

ax2 = fig.add_subplot(132)
ax2.hist(verde.ravel(), 256, [3, 256])
plt.title("Histograma cor verde")

ax3 = fig.add_subplot(133)
ax3.hist(vermelho.ravel(), 256, [3, 253])
plt.title("Histograma cor vermelho")

plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
import cv2

img = cv2.imread("1.png")

#Separa os canais
canais = cv2.split(img)
color  = ("b", "g", "r")
plt.figure()

plt.xlabel("Intensidade")
plt.ylabel("Nmero de Pixels")

for i,col in enumerate(color):
    
    histr = cv2.calcHist([img],[i],None,[256],[1,255])
    plt.plot(histr,color = col)
    plt.xlim([1,256])
    
    ymax = max(histr)
    xmax, _ = np.where(histr == ymax)
    print(ymax, xmax)
    
plt.show()
```
