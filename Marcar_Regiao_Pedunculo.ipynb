```python
import cv2
import json

botao = False
xt, yt = -1, -1  
xb, yb = -1, -1

def marca_area(img, coordenadaInicial, coordenadaFinal, cor, opacidade):
    
    cv2.rectangle(img, coordenadaInicial, coordenadaFinal, cor, int(opacidade))

def area_pedunculo(event, x, y, flags, param):
    
    global xt, yt, xb, yb, botao, img

    if event == cv2.EVENT_LBUTTONDOWN:
        
        botao = True
        xt, yt = x, y
        
    elif event == cv2.EVENT_LBUTTONUP:
        
        botao = False
        xb, yb = x, y
        marca_area(image, (xt, yt), (xb, yb), (0, 255, 0), 1)

    if botao:
        
        image_temp = image.copy()
        marca_area(image_temp, (xt, yt), (x, y), (0, 255, 0), 1)
        cv2.imshow('Pedunculo', image_temp)
        
    else:
        
        cv2.imshow('Pedunculo', image)

cv2.namedWindow('Pedunculo')
cv2.setMouseCallback('Pedunculo', area_pedunculo)

qtdImagem = 50
idx = 11

while(idx <= qtdImagem):
    
    print("TESTE_RES/" + str(idx) + "_ponto_final.jpg")
    
    image = cv2.imread("TESTE_RES/" + str(idx) + "_ponto_final.jpg")
    
    if image is None:
        
        print(f"Erro ao carregar a imagem {idx}_manga_sem_folha.png.")
        
        idx += 1
        
        continue
    
    image = cv2.resize(image, (500, 500))
    
    cv2.imshow('Pedunculo', image)

    key = cv2.waitKey(0)

    if key == 27:
        
        break
        
    if key == ord('p'):
        
        print(xt, yt, xb, yb)
        
        coordinates = {
            'Descricao': {
                'Imagem': str(idx) + "_ponto_final",
                'Objetivo': 'Regiao de Interesse do Pedunculo'
            },
            'Coordenadas':{
                'xt': xt,
                'yt': yt,
                'xb': xb,
                'yb': yb
            }
        }
                
        coordinates_json = json.dumps(coordinates, indent=4)
        
        arquivo = "TESTE_RES/" + str(idx) + "_ponto_final.json"
        
        with open(arquivo, 'w') as file:
            
            file.write(coordinates_json)
            
        xt, yt = -1, -1  
        xb, yb = -1, -1
        
        idx += 1
        
    if key == ord('l'):
        
        xt, yt = -1, -1  
        xb, yb = -1, -1
        
        image = cv2.imread("TESTE_RES/" + str(idx) + "_ponto_final.jpg")

cv2.destroyAllWindows()
```

```python
import json

qtd = 10
i = 1

while(i <= qtd):
    
    imagem = cv2.imread("TESTE_RES/" + str(i) + "_ponto_final.jpg")
    
    if imagem is None:
        
        i += 1
        
        continue
    
    cv2.resize(imagem, (500, 500))
    
    with open('TESTE_RES/' + str(i) + '_ponto_final.json') as file:
    
        coordenadas = json.load(file)
    
    xt, yt = coordenadas['Coordenadas']['xt'], coordenadas['Coordenadas']['yt']
    xb, yb = coordenadas['Coordenadas']['xb'], coordenadas['Coordenadas']['yb']
    
    cv2.rectangle(imagem, (xt, yt), (xb, yb), (255, 0, 0), 1)
        
    cv2.imshow("teste", imagem)
    
    key = cv2.waitKey(0)

    if key == 27:
        
        cv2.imwrite("teste2.png", imagem)
        
        break
        
    if key == ord('p'):
        
        i += 1

cv2.destroyAllWindows()
```
